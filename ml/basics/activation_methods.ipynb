```python
def plot_data(x, y, ax, title, color):
    ax.plot(x, y, c = color) 
    ax.set_xlabel("x") 
    ax.set_ylabel(title) 
    ax.grid()
    ax.set_title(title)
    ax.set_xlim(-5,5)
    ax.set_ylim(0,1)
    ax.set_yticks([0,0.5,1])
    ax.set_xticks([-5,-2.5,0,2.5,5])
    ax.vlines(0, 0, 1, color = "k", linewidth=1)
    ax.hlines(0, -5, 5, color = "k", linewidth=1)
    ax.hlines(0.5, -5, 5, color = "k", linewidth=1)
```

# Activation functions
## Non-linear

### Sigmoid

$ sigmoid(x) =  \frac{\mathrm{1} }{\mathrm{1} + e^{-x} }  $



```python
import matplotlib.pyplot as plt 
import numpy as np 
import math 

def sigmoid(x):
    return 1/(1 + np.exp(-x)) 

def sigmoid_derivative(x):
    return sigmoid(x) * (1 -  sigmoid(x))     

x = np.linspace(-5, 5, 100) 
fig, axs = plt.subplots(1,2, figsize=(8, 3))  
plot_data(x, sigmoid(x), axs[0], "Sigmoid", "r")  
plot_data(x, sigmoid_derivative(x), axs[1], "Sigmoid derivative", "b")   
plt.show() 
```

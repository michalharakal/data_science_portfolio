```python
def plot_data(x, y, ax, title, color):
    ax.plot(x, y, c = color) 
    ax.set_xlabel("x") 
    ax.set_ylabel(title) 
    ax.grid()
    ax.set_title(title)
    ax.set_xlim(min(x),max(x))
    ax.set_ylim(min(y) - 1,max(y) + 1)
    ax.vlines(0, min(x),max(x), color = "k", linewidth=1)
    ax.hlines(0, min(x),max(x), color = "k", linewidth=1)
    #ax.hlines(0.5, -5, 5, color = "k", linewidth=1)
```

# Activation functions
## Non-linear

### Sigmoid

$ sigmoid(x) =  \frac{\mathrm{1} }{\mathrm{1} + e^{-x} }  $

$ sigmoid'(x) =  sigmoid(x) *  (1 - sigmoid(x))  $




```python
import matplotlib.pyplot as plt 
import numpy as np 

def sigmoid(x):
    return 1/(1 + np.exp(-x)) 

def sigmoid_derivative(x):
    return sigmoid(x) * (1 -  sigmoid(x))     

x = np.linspace(-5, 5, 100) 
fig, axs = plt.subplots(1,2, figsize=(8, 3))  
plot_data(x, sigmoid(x), axs[0], "Sigmoid", "r")  
plot_data(x, sigmoid_derivative(x), axs[1], "Sigmoid derivative", "b")   
plt.show() 
```

### Hyperbolic tangent
$tanh(x) = \frac{ e^x - e^{-x} }{e^x + e^{-x} } $

$tanh'(x) = 1 - tanh(x)^2 $



```python
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x)) 

def tanh_derivative(x):
    return 1 - pow(tanh(x), 2)

x = np.linspace(-5, 5, 100) 
fig, axs = plt.subplots(1,2, figsize=(8, 3))  
plot_data(x, tanh(x), axs[0], "tanh", "r")  
plot_data(x, tanh_derivative(x), axs[1], "tanh derivative", "b")   
plt.show() 
```
